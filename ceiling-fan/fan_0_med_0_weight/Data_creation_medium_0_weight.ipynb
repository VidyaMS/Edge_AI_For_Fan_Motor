```python
## no of file - 587
## 200 records each 
## weight = 0
## speed = medium  - 1
```

```python
speed = 1 ## for 'medium'
weight = 0
no_of_files = 587
no_of_chars = 4
```

```python
df = pd.DataFrame()

x = np.empty(1)
y = np.empty(1)
z = np.empty(1)
for i in range(no_of_files):
    file_name = str(i).zfill(no_of_chars) + '.csv'
    data = pd.read_csv(file_name , header = None)
    data.columns = ['x', 'y', 'z']
    data['count'] = i
    df = df.append(data, ignore_index = True)

print(df.shape)
```

```python
df['weight'] =0
df['speed'] = 1  ## for low speed 
```

```python
df.tail()
```

```python
name_of_final_file = 'fan_0_weight_' + str(weight) + '_' + 'speed_' + str(speed) +'.csv'
name_of_final_file
```

```python
## save the file 
df.to_csv(name_of_final_file , index = False)
```
